```python
L = [3,6,5,-6,-7,8,9,5]
for i in range(len(L)):
    for j in range(i+1,len(L)):
        if(L[i]>L[j]):
            temp=L[j]
            L[j]=L[i]
            L[i]=temp
print(L)
```

```python
def fact(x):
    if(x==1):
        return 1
    else:
        return x*fact(x-1)
```
